## Nearest Neighbors Algorithm(unsupervise)

```python
import numpy as np
import pandas as pd
import sklearn
from sklearn.neighbors import NearestNeighbors
```

*mtcars dataset source:* 
Henderson and Velleman (1981), Building multiple regression models interactively. Biometrics, 37, 391â€“411.

```python
cars = pd.read_csv('mtcars.csv')
cars.columns = ['car_names', 'mpg', 'cyl', 'disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb']
cars.head()
```

```python
#select features: mpg,disp,hp,wt 
X = cars.iloc[:,[1, 3, 4, 6]].values
X[0:5]
```

```python
# setup NearestNeighbors values
# mpg,disp,hp,wt
t = [15, 300, 160, 3.2]
t
```

# Import Predictor

```python
# Train model to get 1 most nearest neigbor
nnb = NearestNeighbors(n_neighbors=1)
nbrs = nnb.fit(X)
```

```python
# Calculate most nearest to t[]
print(nbrs.kneighbors([t]))
```

```python
# Get 2 items (first and second) most nearest neigbor
nbrs = NearestNeighbors(n_neighbors=2).fit(X)
print(nbrs.kneighbors([t]))
```
